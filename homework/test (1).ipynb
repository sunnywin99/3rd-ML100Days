```python
#載入所需模組
import numpy as np
import matplotlib.pyplot as plt
```

```python
#設定變數
w=3
b=0.5
```

```python
#利用模組功能，取得x值的陣列
x_lin = np.linspace(0,100,101)

#利用變數方式及陣列的運算，取得y值的陣列
y=(x_lin +np.random.randn(101)*5)*w +b
```

```python
#把xy座標畫出來
plt.plot(x_lin,y,'b.',label = 'data points')
plt.title("Assume we have data points")
plt.legend(loc =2)
plt.show()
```
